# Basic Firewall Configuration (UFW)

This document describes the **minimal firewall configuration** applied to the server using **UFW (Uncomplicated Firewall)**.

The objective is to:
- Allow SSH access for administration
- Block unsolicited incoming traffic
- Keep the configuration simple and safe for a LAN-based server

---

## Firewall Tool

Ubuntu uses **UFW** as a frontend for `iptables`.

Service name:
ufw

```bash
 sudo apt install ufw -y
```

---

## Firewall Status

To check firewall status:

```bash
sudo ufw status
```
Expected output:

```bash
Status: active
```

This confirms that the firewall is enabled and enforcing rules.


```bash
sudo ufw allow ssh
sudo ufw enable
```

Security Notes

+ No ports are opened on the router

+ No port forwarding is configured

+ The firewall only protects the local host

+ SSH access remains limited to the LAN unless explicitly exposed
